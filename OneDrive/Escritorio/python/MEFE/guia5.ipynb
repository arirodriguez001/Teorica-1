```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm, poisson
```

```python
binom_params = [(5, 0.2),(30, 0.4)]
```

```python
for n, p in binom_params:
    mu, sigma = n * p, np.sqrt(n * p * (1 - p))
    x = np.arange(0, n + 1)
    pmf = binom.pmf(x, n, p)
    x_cont = np.linspace(0, n, 500)
    pdf = norm.pdf(x_cont, mu, sigma)

    plt.figure()
    plt.bar(x, pmf, alpha=0.7)
    plt.plot(x_cont, pdf, lw=2)
    plt.title(f'Binomial B({n},{p}) vs Normal\n={mu:.2f}, ={sigma:.2f}')
    plt.xlabel('k')
    plt.ylabel('P(X=k)')
    plt.show()
```

```python
poisson_lams = [4, 10, 40]
for lam in poisson_lams:
    mu, sigma = lam, np.sqrt(lam)
    x = np.arange(0, lam + 4 * int(sigma) + 1)
    pmf = poisson.pmf(x, lam)
    x_cont = np.linspace(0, x.max(), 500)
    pdf = norm.pdf(x_cont, mu, sigma)

    plt.figure()
    plt.bar(x, pmf, alpha=0.7)
    plt.plot(x_cont, pdf, lw=2)
    plt.title(f'Poisson P{lam} vs Normal\n={mu:.1f}, ={sigma: .2f}')
    plt.xlabel('k')
    plt.ylabel('P(X=k)')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

n = 50
Nsim = 10_000

# 1) Genero sumas de 50 uniformes
Z = np.sum(np.random.rand(Nsim, n), axis=1)

# 2) Estandarizo
Y = (Z - n/2) / np.sqrt(n/12)

# 3) Histograma y curva normal
plt.hist(Y, bins=30, density=True, alpha=0.6, label='simulacin')
x = np.linspace(-4, 4, 400)
plt.plot(x, norm.pdf(x, 0, 1), 'r-', lw=2, label='N(0,1)')
plt.xlabel('y')
plt.ylabel('densidad')
plt.legend()
plt.show()
```
