```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
c = 1.0

# Create a grid
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)

# Stream function  = c * x * y
Psi = c * X * Y

# Plot streamlines as contour plot of 
plt.figure()
levels = np.linspace(-4, 4, 17)
contours = plt.contour(X, Y, Psi, levels=levels)
plt.clabel(contours, inline=True, fontsize=8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lneas de corriente para  = c x y')
plt.axhline(0, linewidth=0.5, linestyle='--')
plt.axvline(0, linewidth=0.5, linestyle='--')
plt.show()
```

```python
# Parameter
d = 1.0

# Create grid
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)

# Stream function  = d*(3x^2y - y^3)
Psi = d * (3*X**2 * Y - Y**3)

# Plot streamlines as contour plot of 
plt.figure()
levels = np.linspace(-8, 8, 17)
contours = plt.contour(X, Y, Psi, levels=levels)
plt.clabel(contours, inline=True, fontsize=8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lneas de corriente para  = d(3xy - y)')
plt.axhline(0, linewidth=0.5, linestyle='--')
plt.axvline(0, linewidth=0.5, linestyle='--')
plt.show()
```
