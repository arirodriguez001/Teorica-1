```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
c = 1.0

# Create a grid
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)

# Stream function ψ = c * x * y
Psi = c * X * Y

# Plot streamlines as contour plot of ψ
plt.figure()
levels = np.linspace(-4, 4, 17)
contours = plt.contour(X, Y, Psi, levels=levels)
plt.clabel(contours, inline=True, fontsize=8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Líneas de corriente para ψ = c x y')
plt.axhline(0, linewidth=0.5, linestyle='--')
plt.axvline(0, linewidth=0.5, linestyle='--')
plt.show()
```
